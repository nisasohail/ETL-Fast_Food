```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
income_file = "Resources/kaggle.csv"
income_df = pd.read_csv(income_file, encoding = 'latin-1')
income_df.head()
```

```python
food_file = "Resources/Datafiniti_Fast_Food_Restaurants.csv"
food_df = pd.read_csv(food_file, encoding = 'latin-1')
food_df['postalCode']=food_df['postalCode'].astype(int)
food_df
```

```python
# Create a filtered dataframe from specific columns
income_cols = ["id", "State_ab", "City", "Zip_Code", "Mean", "Median", "Stdev"]
income_transformed = income_df[income_cols].copy()
# Rename the column headers
income_transformed = income_transformed.rename(columns={"id": "id",
                                                       "State_ab": "State",
                                                       "City": "City",
                                                       "Zip_Code": "Zip_Code",
                                                       "Mean": "Mean_Income",
                                                       "Median": "Median_Income",
                                                       "Stdev": "Standard_Deviation_Income"})
# Clean the data by dropping duplicates and setting the index
income_transformed.drop_duplicates("id", inplace=True)
income_transformed.set_index("id", inplace=True)
income_transformed.head()
```

```python
# Create a filtered dataframe from specific columns
food_cols = ["id", "name", "city", "province", "postalCode", "latitude", "longitude"]
food_transformed = food_df[food_cols].copy()
# Rename the column headers
food_transformed = food_transformed.rename(columns={"id": "id",
                                                       "name": "Name",
                                                       "city": "City",
                                                       "province": "State",
                                                       "postalCode": "Zip_Code",
                                                       "latitude": "Latitude",
                                                       "longitude": "Longitude"})
# Clean the data by dropping duplicates and setting the index
food_transformed.drop_duplicates("id", inplace=True)
food_transformed.set_index("id", inplace=True)
food_transformed.head()
```

```python
combined =pd.merge(income_transformed, food_transformed, on='Zip_Code')
combined
```

```python
combined_cols = ["Name","City_x", "State_x","Zip_Code", "Latitude", "Longitude" , "Mean_Income", "Median_Income",
       "Standard_Deviation_Income"]
combined_transformed = combined[combined_cols].copy()
# Rename the column headers
combined_transformed = combined_transformed.rename(columns={"City_x":"City",
                                                        "State_x":"State"})
combined_transformed.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/food_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
combined_transformed.to_sql(name='fast_food', con=engine, if_exists='append', index=True)
```
